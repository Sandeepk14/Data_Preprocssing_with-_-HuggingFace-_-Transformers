# Tokenization with HuggingFace Transformers

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, AutoModelForSequenceClassification, DistilBertTokenizer, DistilBertForSequenceClassification
```

```python
from transformers import DistilBertTokenizer
```

```python
tokenizer= DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
```

```python
text= "Hello, My name is Sandeep Kumar, I am a AI Enginner."
```

```python
tokens= tokenizer.tokenize(text)
token_id= tokenizer.convert_tokens_to_ids(tokens)
```

```python
print("Tokens:", tokens)
print("Token IDs:", token_id)
```

# Tokenization with HuggingFace Transformers Model Gpt2tokenizer

```python
from transformers import GPT2Tokenizer
tokenizer= GPT2Tokenizer.from_pretrained('gpt2')
text= "Hello, My name is Sandeep Kumar, I am a AI Enginner."
tokens= tokenizer.tokenize(text)
token_id= tokenizer.convert_tokens_to_ids(tokens)
print("Tokens:", tokens)
print("Token IDs:", token_id)
```

```python
from transformers import BertTokenizer
# load the BERT Tokenizer
tokenizer= BertTokenizer.from_pretrained('bert-base-uncased')

# tokenize a sample text
text= "hello, my name is Sandeep Kumar, I am a AI Enginner."
tokens= tokenizer.tokenize(text)
token_id= tokenizer.convert_tokens_to_ids(tokens)

# print the results
print("Tokens:", tokens)
print("Token IDs:", token_id)
print("Encoded:", tokenizer.encode_plus(text,return_tensors='pt'))
```

```python
from transformers import BertTokenizer
# load the BERT Tokenizer
tokenizer= BertTokenizer.from_pretrained('bert-base-uncased')

# Exampe a input sentence
sentence1 = "hello, how are you?"
sentence2 =  "I am fine, thank you."

# Tokenize the sentences with special tokens
encoded = tokenizer.encode_plus(
    sentence1,
    sentence2,
    add_special_tokens=True, # Add [cls], [sep] tokens
    max_length=128,          # Max length of the sentence
    padding='max_length',    # Pad to fix length
    truncation=True,         # Truncate if input exceeds max length
    return_attention_mask=True, # Return Attention Mask of each sentence
    return_tensors='pt'      # Return Pytorch Tensors
)

# print the encoded outputes
print("Encoded Input IDs", encoded['input_ids'])
print("Token Type IDs",encoded['token_type_ids'])
print("Attention Mask", encoded['attention_mask'])

# Decode the input IDs back to the original santences
decoded= tokenizer.decode(encoded['input_ids'][0],skip_special_tokens=False)
print("Decoded Sentence:", decoded)

```
